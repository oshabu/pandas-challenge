```python
# Combine the data into a single dataset.  
merge_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
merge_df.head()
```

# District Summary
Calculate the total number of schools
Calculate the total number of students
Calculate the total budget
Calculate the average math score
Calculate the average reading score
Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
Calculate the percentage of students with a passing math score (70 or greater)
Calculate the percentage of students with a passing reading score (70 or greater)
Create a dataframe to hold the above results
Optional: give the displayed data cleaner formatting

```python
#District Summary
unique_school_names = schools_df['school_name'].unique()
school_count = len(unique_school_names)

total_students = schools_df['size'].sum()
total_budget = schools_df['budget'].sum()

#number and % passing reading
passing_reading = students_df.loc[students_df['reading_score'] >= 70]['reading_score'].count()
perc_pass_reading = (passing_reading/total_students) * 100
perc_pass_reading=round(perc_pass_reading,2)
#number and % passing math
passing_math = students_df.loc[students_df['math_score'] >= 70]['math_score'].count()
perc_pass_math = (passing_math/total_students) * 100
perc_pass_math = round(perc_pass_math,2)
#average scores
avg_math_score = students_df['math_score'].mean()
avg_math_score=round(avg_math_score,2)
avg_reading_score = students_df['reading_score'].mean()
avg_reading_score=round(avg_reading_score,2)
overall_passing = (perc_pass_math + perc_pass_reading) / 2

district_summary = pd.DataFrame({"Total Schools": [school_count],
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "Average Reading Score": [avg_reading_score],
                                 "Average Math Score": [avg_math_score],
                                 "% Passing Reading":[perc_pass_reading],
                                 "% Passing Math": [perc_pass_math],
                                 "Overall Passing Rate": [overall_passing]})
# Reformat Total Students (with comma)and Total Budget  (with comma and $)
district_summary['Total Students'] = district_summary.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
district_summary['Total Budget'] = district_summary.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
district_summary
```

# School Summary
Create an overview table that summarizes key metrics about each school, including:

School Name
School Type
Total Students
Total School Budget
Per Student Budget
Average Math Score
Average Reading Score
% Passing Math
% Passing Reading
Overall Passing Rate (Average of the above two)
Create a dataframe to hold the above results

```python
#School Summary
by_school = merge_df.set_index('school_name').groupby(['school_name'])
sch_types = schools_df.set_index('school_name')['type']
stu_per_sch = by_school['Student ID'].count()

sch_budget = schools_df.set_index('school_name')['budget']
stu_budget = schools_df.set_index('school_name')['budget']/schools_df.set_index('school_name')['size']

avg_math = by_school['math_score'].mean()
avg_math = round(avg_math,2)
avg_read = by_school['reading_score'].mean()
avg_read=round(avg_read,2)

# % passing
pass_math = merge_df[merge_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch * 100
pass_math = round(pass_math,2)
pass_read = merge_df[merge_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch * 100
pass_read = round(pass_read,2)
overall = (pass_math + pass_read) / 2
overall = round(overall,2)
school_summary = pd.DataFrame({
    "School Type": sch_types,
    "Total Students": stu_per_sch,
    "Per Student Budget": stu_budget,
    "Total School Budget": sch_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "% Overall Passing Rate": overall})
#reformat Total Students (with comma) and Total school Budget  (with comma and $)
school_summary['Per Student Budget'] = school_summary.apply(lambda x: "${:,.2f}".format(x['Per Student Budget']), axis=1)
school_summary['Total Students'] = school_summary.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
school_summary['Total School Budget'] = school_summary.apply(lambda x: "${:,.2f}".format(x['Total School Budget']), axis=1)
school_summary
```

# Scores by School Spending
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
Average Math Score
Average Reading Score
% Passing Math
% Passing Reading
Overall Passing Rate (Average of the above two)

```python
#Scores by school spending
bins = [0, 580, 610, 640, 6000]
group_name = ['< $580', "$580 - 610", "$610 - 640", "> $640"]

merge_df['budget_bins'] = pd.cut(merge_df['budget']/merge_df['size'], bins, labels = group_name)
grouped_spending = merge_df.groupby('budget_bins')

avg_math = grouped_spending['math_score'].mean()
avg_math=round(avg_math,2)
avg_read = grouped_spending['reading_score'].mean()
avg_read=round(avg_read,2)
pass_math = merge_df[merge_df['math_score'] >= 70].groupby('budget_bins')['Student ID'].count()/grouped_spending['Student ID'].count() * 100
pass_math=round(pass_math,2)
pass_read = merge_df[merge_df['reading_score'] >= 70].groupby('budget_bins')['Student ID'].count()/grouped_spending['Student ID'].count() * 100
pass_read=round(pass_read,2)
overall = merge_df[(merge_df['reading_score'] >= 70) & (merge_df['math_score'] >= 70)].groupby('budget_bins')['Student ID'].count()/grouped_spending['Student ID'].count() * 100
overall=round(overall,2)           
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "% Overall Passing Rate": overall})
scores_by_spend
```

# Scores by School Size

```python
#Scores by School Size
bins = [0, 799, 2399, 23000]
group_names = ["Small (<800)", "Medium (800-2400)" , "Large (>2400)"]
merge_df['size_bins'] = pd.cut(merge_df['size'], bins, labels = group_names)

by_size = merge_df.groupby('size_bins')
 
avg_math = by_size['math_score'].mean()
avg_math=round(avg_math,2)
avg_read = by_size['math_score'].mean()
avg_read=round(avg_read,2)
pass_math = merge_df[merge_df['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count() * 100
pass_math=round(pass_math,2)
pass_read = merge_df[merge_df['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count() * 100
pass_read=round(pass_read,2)
overall = merge_df[(merge_df['reading_score'] >= 70) & (merge_df['math_score'] >= 70)].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count() * 100
overall=round(overall,2)          
scores_by_size = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "% Overall Passing Rate": overall
})          
scores_by_size
```
